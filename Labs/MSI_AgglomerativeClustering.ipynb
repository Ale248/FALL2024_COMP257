```python
#import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn import metrics
```

```python
#Load the iris dataset
iris = datasets.load_iris()
X= iris.data
y=iris.target
```

```python
X, y
```

```python
X.shape
```

```python
linkage_used = "ward"
metric_used = "euclidean"
#Build the classifier
iris_clf = AgglomerativeClustering(n_clusters=3, metric=metric_used, linkage=linkage_used)

iris_clf.fit(X)
data_labels = iris_clf.labels_

print("Data Labels:" ,data_labels)

print("Silhouette Score:" ,metrics.silhouette_score(X, data_labels))
```

```python
data_labels.shape
```

```python
# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=data_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(20, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Labels')
plt.ylabel('Distance')
dendro = sch.dendrogram(sch.linkage(X,linkage_used, metric=metric_used),
                        labels=data_labels,
                        leaf_rotation=90,
                        leaf_font_size=8,
                        show_contracted=True)
plt.show()
```

```python
#Build the classifier

iris_clf_3 = AgglomerativeClustering(n_clusters=3, metric=metric_used, linkage=linkage_used)

iris_clf_3.fit(X)
data_labels_3 = iris_clf_3.labels_

print("Data Labels:" ,data_labels_3)

print("Silhouette Score:" ,metrics.silhouette_score(X, data_labels_3))
```

```python
# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=data_labels_3, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering')
plt.show()
```

```python
# Plot the dendrogram
plt.figure(figsize=(20, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Labels')
plt.ylabel('Distance')
sch.set_link_color_palette(['m', 'c', 'y', 'k'])
dendro = sch.dendrogram(sch.linkage(X, linkage_used, metric=metric_used),
                        labels=data_labels_3,
                        leaf_rotation=90,
                        leaf_font_size=8,
                        show_contracted=True)
plt.show()
```
